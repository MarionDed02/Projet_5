## Univariate analysis

### Quantitative variable

Let's perform a univariate analysis of the numerical variables and calculate for each variable :
- its mean
- its median
- its standard deviation
- its variance
- its minimum
- its maximum
- its first quartile
- its third quartile
- empirical skewness and empirical kurtosis to capture shape measures
- and represent its distribution with a histogram.

```python
# Select numeric columns only
numeric_columns = data.select_dtypes(include=[np.number])

# Iterate with numeric columns
for colonne in numeric_columns.columns:
    # Statistics calculation
    moyenne = numeric_columns[colonne].mean()
    mediane = numeric_columns[colonne].median()
    ecart_type = numeric_columns[colonne].std()
    variance = numeric_columns[colonne].var()
    minimum = numeric_columns[colonne].min()
    maximum = numeric_columns[colonne].max()
    quantile_25 = numeric_columns[colonne].quantile(0.25)
    quantile_75 = numeric_columns[colonne].quantile(0.75)
    skewness = numeric_columns[colonne].skew()
    kurt = numeric_columns[colonne].kurtosis()
    

    # Displaying statistics
    print(f"Statistics for '{colonne}':")
    print(f"Mean: {moyenne:.2f}")
    print(f"Median: {mediane:.2f}")
    print(f"Standard deviation: {ecart_type:.2f}")
    print(f"Variance: {variance:.2f}")
    print(f"Minimum: {minimum:.2f}")
    print(f"Maximum: {maximum:.2f}")
    print(f"Quartile 25%: {quantile_25:.2f}")
    print(f"Quartile 75%: {quantile_75:.2f}")
    print(f"Skewness: {skewness:.2f}")
    print(f"Kurtosis: {kurt:.2f}")   

    # Creating the histogram
    plt.figure(figsize=(5, 3))
    sns.histplot(data=data, x=colonne, bins=20, kde=False, color='lightblue', edgecolor='k')
    plt.title(f' "{colonne}" - Histogram', fontsize=12)
    plt.xlabel('Values', fontsize=10)
    plt.ylabel('Frequency', fontsize=10)
    plt.tick_params(axis='both', labelsize=10)
    plt.show()
    
    # Creating the box plot
    plt.figure(figsize=(8, 3))
    sns.boxplot(data=data, x=colonne, color='lightgreen', orient='h')
    plt.title(f' "{colonne}" - Boxplot', fontsize=12)
    plt.xlabel('Values', fontsize=10)
    plt.tick_params(axis='both', labelsize=10)
    plt.show()
    

    print("-------------------------------------------------------------------------------------------")
    print(' ')
    print(' ')
```

# Features selection

## Creation of RFM features

We will create three new variables for our RFM analysis:
- **Recency**: The "Recency" variable measures the duration since a customer's last order. The lower the recency value, the more recent the customer is considered.
     - We find the most recent date in our data (the highest date in 'order_purchase_timestamp') and add two days to it.
     - We calculate the difference between this most recent date and the purchase date of each order for each customer in days.
- **Frequency**: The “Frequency” variable measures the number of orders placed by a customer.
     - We count the total number of orders for each customer.
- **Monetary**: The "Monetary" variable measures the cumulative amount of orders placed by a customer.
     - Calculation of the total sum of amounts (prices) of all customer orders.

```python
# maximum date in the 'order_purchase_timestamp' column
maximum_date = data['order_purchase_timestamp'].max()

# Added 2 days to the maximum date for recency
date_reference = maximum_date + pd.Timedelta(days=2)

# Calculation of recency in days
recency = (date_reference - data.groupby('customer_unique_id')['order_purchase_timestamp'].max()).dt.days


frequency = data.groupby('customer_unique_id')['order_id'].nunique()

monetary = data.groupby('customer_unique_id')['payment_value'].sum()

# Created a new DataFrame for RFM variables.
rfm_df = pd.DataFrame({
     'recency': recency,
     'frequency': frequency,
     'monetary': monetary
})
```

Let's check for outliers that fall within plus or minus 5%:

```python
def num_outlier(df_in, col_name):
    q1 = df_in[col_name].quantile(0.05)
    q3 = df_in[col_name].quantile(0.95)
    iqr = q3-q1 
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    outliers_df= df_in.loc[(df_in[col_name] < fence_low) | (df_in[col_name] > fence_high)]
    return print("Number of outliers in {} column : ".format(col_name), len(outliers_df)), print("Index: ", outliers_df.index)
```

```python
for i in ["recency", "frequency", "monetary"]:
    num_outlier(rfm_df, i)
    print("*"*40)
```

```python
rfm_df.shape
```

Let's remove the outliers from the Monetary column:

```python
def remove_outlier(df_in, col_name):
    q1 = df_in[col_name].quantile(0.05)
    q3 = df_in[col_name].quantile(0.95)
    iqr = q3-q1     
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    index_outliers= df_in.loc[(df_in[col_name] < fence_low) | (df_in[col_name] > fence_high)].index
    df_in= pd.DataFrame(df_in.drop(index_outliers.to_list(), axis=0, inplace=True))
    print("The outliers in the {} column have been removed".format(col_name))
    return df_in
```

```python
for i in ["monetary"]:
    remove_outlier(rfm_df, i)
    print("*"*40)
```
